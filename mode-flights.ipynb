```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('flights_dataset.csv')
```

```python
df = df.fillna(np.nan)
```

```python
df.sample(n=5)
```

```python
df.sort_values(by='arr_delay', ascending=False)[:20]
```

```python
df['delayed'] = df['arr_delay'].apply(lambda x: x > 0)
```

```python
df['delayed'].value_counts()
```

```python
cancelled = df['cancelled'].value_counts()[1]
on_time = df['delayed'].value_counts()[0]
delayed = df['delayed'].value_counts()[1]
total = on_time + delayed 
print(f'On-time pct: {on_time / total}')
print(f'Delayed pct: {delayed / total}')
```

```python
group_by_carrier = df.groupby(['unique_carrier', 'delayed'])
```

```python
group_by_carrier.size()
```

```python
count_delays_by_carrier = group_by_carrier.size().unstack()
count_delays_by_carrier
```

```python
count_delays_by_carrier.plot(kind='barh', stacked=True, figsize=[16,6], colormap='winter')
```

```python
# How many flights were delayed longer than 20 mins?
# What pct of total flights does this represent?
# What pct of delayed flights does this represent?
```

```python
df['over_20_min_delay'] = df['arr_delay'].apply(lambda x: x > 20)
df.groupby('over_20_min_delay').size()
```

```python
count_over_20_delayed = df.groupby('over_20_min_delay').size()[1]

delays_over_20_mins_vs_total_flights = count_over_20_delayed / len(df)
delays_over_20_mins_vs_total_flights
```

```python
delays_over_20_mins_vs_all_delayed = count_over_20_delayed / delayed 
delays_over_20_mins_vs_all_delayed
```

```python
df['delayed'].head()
```

```python
delayed_flights = df[df['delayed'] == True]
delayed_flights['unique_carrier'].value_counts()
```

```python
delayed_count_by_carrier = df.groupby(['unique_carrier', 'delayed']).size().unstack().reset_index()
delayed_count_by_carrier[:5]
```

```python
delayed_count_by_carrier['total_flights'] = delayed_count_by_carrier[False] + delayed_count_by_carrier[True]
delayed_count_by_carrier[:5]
```

```python
delayed_count_by_carrier['pct_delayed'] = delayed_count_by_carrier[True] / delayed_count_by_carrier['total_flights']
delayed_count_by_carrier[:5]
```

```python
delayed_count_by_carrier.sort_values('pct_delayed', ascending=False)
```
